```python
sns.relplot(
    data = titanic, 
    x= "age" , y = "fare",
    alpha= 0.5, hue = "sex",
    col ="survived"
)
```

```python
sns.relplot(
    data = titanic, 
    x= "age" , y = "fare",
    alpha= 0.5, hue = "sex",
    col ="survived",
    palette = {"male" : "blue", "female" : "magenta"}
)
```

```python
titanic['Sex']= titanic['sex'].map({"male":"Male","female": "Female"})
sns.relplot(
    data = titanic, 
    x= "age" , y = "fare",
    alpha= 0.5, hue = "sex",
    col ="survived",
    palette = {"male" : "blue", "female" : "red"}
)
```

```python
sns.relplot(
    data = titanic, 
    x= "age" , y = "fare",
    alpha= 0.5, hue = "sex",
    col ="survived",size= "pclass",
    palette = {"male" : "#5e78a2", "female" : "#bd2a9e"}
    
)
```

```python
titanic['survived_ind'] = titanic['survived'].map({0:"didn't survived", 1: "survived"})
sns.relplot(
    data = titanic, 
    x= "age" , y = "fare",
    alpha= 0.5, hue = "sex",
    col ="survived_ind",size= "pclass",
    palette = {"male" : "#5e78a2", "female" : "#bd2a9e"}
    
)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
smh = pd.read_csv(r"C:\Users\Agast\Downloads\archive (1)\Student Mental health.csv")
smh.head()
```

```python
smh.groupby("Choose your gender")[['Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?']].count()
```

```python
columns_to_plot = smh.columns[-4: -1]

category_order = ['Yes','No'] 
```

```python
for column in columns_to_plot:
    counts = smh[column].value_counts()
    counts = counts.reindex(category_order, fill_value=0)
    
    print(counts)
    print()
```

```python
sns.countplot(smh, x= 'Do you have Depression?', stat='percent', hue = 'Did you seek any specialist for a treatment?')
```
